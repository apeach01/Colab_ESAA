## **04-1** 데이터 시각화가 필요한 이유
### * 앤스콤 함정 및 데이터 시각화 필요성
  - **앤스콤 4분할 그래프 구성 데이터 집합:** 4개의 그룹으로 되어 있어 모든 데이터 그룹이 x, y열을 가짐

## **- 다변량 그래프 그리기**
  - 다변량 그래프: 3개 이상의 변수를 사용한 그래프
  - 적재적소에 맞는 그래프 요소를 추가하여 그래프를 표현하기 위해 다변량 그래프 필요함

#### 1. 다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프 
#### 1) 앞실습에서 한 그래프+성별 변수 추가하기
  - 산점도 그래프에서 색상으로 구분하면 됨
  - 문자열의 변수는 색상 지정값으로 사용할 수 없음
  - 따라서 0,1과 같은 정수를 사용해야 함
  - 예) 남자 = 1, 여자 = 0

#### 2) recode_sex 메서드가 반환한 값을 dataframe에 추가하기
  - recode_sex 메서드가 반환한 값을 새로운 열을 생성해 추가해야함
  - 브로드캐스팅하기 위해 apply 메서드 사용함

#### 3) 인원수 변수도 그래프에 추가하기
  - 테이블 인원수는 점의 크기로 표현하면 됨
  - scatter 메서드에 s,c 인잣값으로 테이블당 인원 수 및 성별의 치환값을 전달하면 됨

```python
# 1
def recode_sex(sex) : 
  if sex == 'Female' :
    return 0
  else : 
    return 1
```

```python
# 2
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
# 3
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size']*10,
    c = tips['sex_color'],
    alpha = 0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

  - 통계 수치 확인을 위해 자주 쓰임
  - 데이터 분산이 모호하게 표현됨 -> 커널 밀도를 추정한 바이올린 그래프 사용하면 됨 -> violinplot 메서드 사용

#### 7) 관계 그래프 그리기
  - 위에서 언급한 모든 그래프를 종합한 그래프
  - pairplot 사용함
  - 중복된 정보가 표현될 수 있는 단점
  - 관계 그래프의 절반은 나머지 절반과 같음
  - 해결: 중복된 그래프가 그려지는 위치를 직접 지정하여 원하는 그래프로 교체함
  - map_upper: 대각선을 기준으로 위쪽에 그릴 그래프 지정
  - map_lower: 대각선을 기준으로 아래쪽에 그릴 그래프 지정
  - map_diag: 대각선 중심으로 그래프 그리기

```python

# 1
ax = plt.subplots()
ax = sns.regplot(x= 'total_bill', y = 'tip', data = tips)
ax.set_title('Scatterplot of Total Bill and Trip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

ax = plt.subplots()
ax = sns.regplot(x= 'total_bill', y = 'tip', data = tips, fit_reg = False)
ax.set_title('Scatterplot of Total Bill and Trip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 2
joint = sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize = 10, y = 1.03)
```

```python
# 3
hexbin = sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
hexbin.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize = 10, y =1.03)
```

```python
# 4
ax = plt.subplots()
ax = sns.kdeplot(data = tips['total_bill'],
                 data2 = tips['tip'],
                 shade = True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 5
ax = plt.subplots()
ax = sns.barplot(x = 'time', y = 'total_bill', data = tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
# 6
ax = plt.subplots()
ax = sns.boxplot(x = 'time', y = 'total_bill', data = tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

```python
# 7
ax = plt.subplots()
ax = sns.violinplot(x = 'time', y = 'total_bill', data = tips)
ax.set_title('Violin plot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

```python
# 8
fig = sns.pairplot(tips)
```

```python

# 9
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug = True)
plt.show()
```
